```python
import gym
import PIL
from IPython import display
import numpy as np
from IPython.display import clear_output
import time
import sys
```

```python
action_n=env.action_space.n
observation_n=env.observation_space.n
```

```python
qtable = np.zeros((observation_n, action_n))
print(qtable)
```

```python
total_episodes = 15000
learning_rate = 0.8
max_steps_in_episode = 99
gamma = 0.9
# Exploration parameters
epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.1 
decay_rate = 0.005   
```

```python
rewards=[]
for episode in range(total_episodes):
    # Reset the environment
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    
    print('episode {}'.format(episode))
    for step in range(max_steps_in_episode):
        random_no = np.random.uniform(0, 1)
        
        ## exploitation (taking the biggest Q value for this state)
        env.render()
        if random_no > epsilon:
            action = np.argmax(qtable[state,:])

        # exploration
        else:
            action = env.action_space.sample()
        # continue with the new state
        observation, reward, done, info = env.step(action)

        # Update Q(s,a):= Q(s,a) + lr [R(s,a) + gamma * max Q(s',a') - Q(s,a)]
        # qtable[new_state,:] : all the actions we can take from new state
        qtable[state, action] = qtable[state, action] + learning_rate * (reward + gamma * np.max(qtable[observation, :]) - qtable[state, action])
        
        total_rewards += reward
        
        # Our new state is state
        state = observation
        
        # If done (if we're dead) : finish episode
        if done == True: 
            break
    if episode%5==0:
        clear_output()  
    # Reduce epsilon (because we need less and less exploration)
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode) 
    rewards.append(total_rewards)

print ("Score over time: {}".format(sum(rewards)/total_episodes))
print(qtable)
```
